### 2.2 More Data Manipulation Functions
#### 2.2.1 Filtering Data

````python
import pandas as pd

# Original DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['New York', 'San Francisco', 'Los Angeles']
})

# Additional data
additional_data = {
    'Name': ['David', 'Emily', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack', 'Katie', 'Leo', 'Mia','Alex','Yasser'],
    'Age': [28, 35, 40, 26, 33, 29, 27, 31, 36, 24,28 ,np.NaN],
    'City': ['Chicago', 'Seattle', 'Miami', 'Denver', 'Boston', 'Phoenix', 'Austin', 'Portland', 'Atlanta', 'Berlin','Meschede','Dallas']
}

# Convert additional data to DataFrame
additional_df = pd.DataFrame(additional_data)

# Concatenate the original DataFrame with the additional data
extended_df = pd.concat([df, additional_df], ignore_index=True)

print(extended_df)
df = extended_df
df
````

````python
#Filtering data based on a condition
filtered_df = df[df['Age'] > 25]
filtered_df
````

````python
# Filtering data based on a condition
filtered_df = df[df['Age'] > 16]
filtered_df
````
